```python
from transformers import AutoTokenizer

model = "teknium/Mistral-Trismegistus-7B"
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
from transformers import pipeline
import torch

llama_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.float16,
    trust_remote_code=True,
    device_map="auto",
)
```

```python
llama_pipeline('hi')
```

```python
import gradio as gr

def yes_man(message, history):
    return  llama_pipeline(message, max_length=500)[0]['generated_text']

gr.ChatInterface(
    yes_man,
    chatbot=gr.Chatbot(height=800),
    textbox=gr.Textbox(placeholder="Ask me question about spirituality", container=False, scale=7),
    title="Yes Man",
    description="Ask me question about spirituality",
    theme="soft",
    examples=["What is life"],
    cache_examples=True,
    retry_btn=None,
    undo_btn="Delete Previous",
    clear_btn="Clear",
).launch(share=True)
```
